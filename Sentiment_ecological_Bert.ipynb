## Summary in one .csv for all reports

```python
import pandas as pd
company_list_names_year=[]
#Clean and import the 10K reports (here from 2019 to 2022)
#To change if want to import something from another file:
company_list_dir = os.getcwd() + '/computation_ecological_bert'
#To keep if want to import from another file:
company_list_csv = os.listdir(company_list_dir)
for i in company_list_csv:
    company_list_names_year=company_list_names_year+[i[4:-4]]
count = 0
df=pd.DataFrame()
for i in company_list_csv:
    path=company_list_dir + '/'+ i
    df_tempo = pd.read_csv(path)
    percentage_esg_sentences=sum(df_tempo['classification'])/len(df_tempo)*100
    esg_score=percentage_esg_sentences/100 - (1-percentage_esg_sentences/100)
    df_results=pd.DataFrame(np.array([[company_list_names_year[count], percentage_esg_sentences, esg_score]]),
                   columns=['company_date', 'percentage_esg_sentences', 'esg_score'])
    df=pd.concat([df, df_results], ignore_index=True)
    count=count+1
# save to outfile
df.to_csv('Sentiment_Ecological_Bert', index=False)
```
