```python
from PyPDF2 import PdfReader
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'Sentimental_Bert_on_10K.csv')
ESG_scores_MSCI = pd.read_csv('msci scores - Msci_scores(1).csv', delimiter=',')
#ESG_scores_MSCI = pd.read_excel('msci scores - Msci_scores.xlsx')
```

```python
#sigmoid function
def sigmoid(x) :
    y = 1/(1 + np.exp(-x))
    return y
    
```

```python
# positive values 
df['average_compound']= df['average_compound']+1
#df['average_compound'] = sigmoid(df['average_compound'])
```

```python
#BAC
x = [df['average_compound'][0],df['average_compound'][1],df['average_compound'][2]]
y = [ESG_scores_MSCI['2019'][8],ESG_scores_MSCI['2020'][8],ESG_scores_MSCI['2021'][8]]

slope, intercept, r, p, std_err = stats.linregress(x, y)

regsummary= sm.OLS(y,x).fit()
regsummary.summary()

#with open('summary_BAC_BERT_MSCI.csv', 'w') as fh:
    #fh.write(regsummary.summary().as_csv())
```

```python
# Regression of all companies on Y : esg scores of 2019
dataset = pd.read_csv('Data_set_for_regression_BERT_MSCI.csv', delimiter=';')

y = dataset[['2019_msci']]
x = dataset[['2019_compound']]

#add constant to predictor variables
x = sm.add_constant(x)

#fit linear regression model
model = sm.OLS(y, x).fit()

model.summary()

#with open('summary_allcompanies_BERT_MSCI_2019.csv', 'w') as fh:
    #fh.write(model.summary().as_csv())
```
